# Linear Regression Model

### Importing the necessary packages for linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X=wine.drop(['quality'],axis=1)
y=wine['quality']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size= .30,random_state=26)
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
# this gives the co-efficent values of linear regression
```

```python
model.intercept_
# this gives the intercept value, nothing but b0(beta-not) value
```

### Every column and their each respective Co-efficient

```python
c_dataframe=pd.DataFrame(model.coef_, X.columns, columns=['coefficients'])
c_dataframe
```

### Predicted values of the model

```python
predictions=model.predict(X_test)
predictions
```

# Evaluating the Regression Model

### Importing the necessary packages for evaluating the linear regression model

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
MAE = metrics.mean_absolute_error(y_test,predictions)
MAE
# gives the mean absolute error value
```

```python
MSE = metrics.mean_squared_error(y_test,predictions)
MSE
# gives the mean squared error value
```

```python
RMSE = np.sqrt(MSE)
RMSE
# gives the root mean squared error value
```

```python
r2 = metrics.r2_score(y_test,predictions)
r2
# Gives the R-Square value
```
